```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 8]

# --- Dataset ---
X = X_df.values

# --- Shuffle e split ---
X_shuffled, y_shuffled = shuffle(X, y, random_state=42)

X_normal = X_shuffled[y_shuffled == 0]
X_anomaly = X_shuffled[y_shuffled == 1]

n_normal = X_normal.shape[0]
n_anomaly = X_anomaly.shape[0]

train_size = int(0.8 * n_normal)
cv_normal_size = int(0.1 * n_normal)
cv_anomaly_size = int(0.5 * n_anomaly)

X_train_raw = X_normal[:train_size]  # solo normali
X_cv_normal = X_normal[train_size:train_size + cv_normal_size]
X_cv_anomaly = X_anomaly[:cv_anomaly_size]
X_cross_val_raw = np.vstack((X_cv_normal, X_cv_anomaly))
y_cross_val = np.hstack((np.zeros(cv_normal_size), np.ones(cv_anomaly_size)))

X_test_normal = X_normal[train_size + cv_normal_size:]
X_test_anomaly = X_anomaly[cv_anomaly_size:]
X_test_raw = np.vstack((X_test_normal, X_test_anomaly))
y_test = np.hstack((np.zeros(len(X_test_normal)), np.ones(len(X_test_anomaly))))

# --- Scaler MinMax per output sigmoid ---
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train_raw)
X_cross_val = scaler.transform(X_cross_val_raw)
X_test = scaler.transform(X_test_raw)

print(f"Training set size: {X_train.shape[0]} (all normal)")
print(f"Cross-validation set size: {X_cross_val.shape[0]} ({cv_normal_size} normal, {cv_anomaly_size} anomalies)")
print(f"Test set size: {X_test.shape[0]} ({len(X_test_normal)} normal, {len(X_test_anomaly)} anomalies)")

LATENT_DIM = 32  # Aumentato da 16 a 32
INPUT_DIM = X_train.shape[1]

# --- Generator migliorato con BatchNorm e architettura pi complessa ---
def build_generator():
    model = models.Sequential([
        layers.Input(shape=(LATENT_DIM,)),

        # Primo blocco
        layers.Dense(128),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),

        # Secondo blocco
        layers.Dense(256),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),

        # Terzo blocco
        layers.Dense(512),
        layers.BatchNormalization(),
        layers.LeakyReLU(alpha=0.2),

        # Output layer
        layers.Dense(INPUT_DIM, activation='sigmoid'),
    ])
    return model

# --- Discriminator migliorato con Dropout e BatchNorm ---
def build_discriminator():
    inp = layers.Input(shape=(INPUT_DIM,))

    # Primo blocco
    x = layers.Dense(512)(inp)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(alpha=0.2)(x)
    x = layers.Dropout(0.3)(x)

    # Secondo blocco
    x = layers.Dense(256)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(alpha=0.2)(x)
    x = layers.Dropout(0.3)(x)

    # Terzo blocco
    x = layers.Dense(128)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(alpha=0.2)(x)
    x = layers.Dropout(0.3)(x)

    # Output layer
    out = layers.Dense(1, activation='sigmoid')(x)

    return models.Model(inp, out)

# --- Callback per Early Stopping ---
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='g_loss',
    patience=100,
    restore_best_weights=True
)

# --- Funzione per salvare i pesi del modello ---
checkpoint = tf.keras.callbacks.ModelCheckpoint(
    filepath='best_gan_model.h5',
    save_best_only=True,
    monitor='g_loss',
    mode='min'
)

generator = build_generator()
discriminator = build_discriminator()

# --- Learning rate scheduler ---
def lr_scheduler(epoch, lr):
    if epoch > 2000:
        return lr * 0.98
    return lr

lr_callback = tf.keras.callbacks.LearningRateScheduler(lr_scheduler)

def train_gan(epochs=5000, batch_size=128):  # Aumentato epoche e batch size
    optimizer_d = tf.keras.optimizers.AdamW(learning_rate=0.0002, beta_1=0.5, weight_decay=1e-4)
    optimizer_g = tf.keras.optimizers.AdamW(learning_rate=0.0001, beta_1=0.5, weight_decay=1e-4)

    discriminator.compile(loss='binary_crossentropy', optimizer=optimizer_d)
    discriminator.trainable = False

    z = layers.Input(shape=(LATENT_DIM,))
    fake = generator(z)
    valid = discriminator(fake)
    combined = models.Model(z, valid)
    combined.compile(loss='binary_crossentropy', optimizer=optimizer_g)

    history = {
        'd_loss': [],
        'g_loss': []
    }

    for epoch in range(epochs):
        # --- Training Discriminator ---
        discriminator.trainable = True

        # Seleziona un batch casuale di esempi
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        real_data = X_train[idx]

        # Genera un batch di nuovi dati
        z_noise = np.random.normal(0, 1, (batch_size, LATENT_DIM))
        gen_data = generator.predict(z_noise, verbose=0)

        # Label smoothing
        real_labels = np.ones((batch_size, 1)) * 0.9 + 0.1 * np.random.rand(batch_size, 1)
        fake_labels = np.zeros((batch_size, 1)) + 0.1 * np.random.rand(batch_size, 1)

        # Addestra il discriminatore
        d_loss_real = discriminator.train_on_batch(real_data, real_labels)
        d_loss_fake = discriminator.train_on_batch(gen_data, fake_labels)
        d_loss = 0.5 * (d_loss_real + d_loss_fake)

        # --- Training Generator ---
        discriminator.trainable = False

        # Campione nuovo rumore per il generatore
        z_noise = np.random.normal(0, 1, (batch_size, LATENT_DIM))

        # Label per il training del generatore
        valid_labels = np.ones((batch_size, 1))

        # Addestra il generatore
        g_loss = combined.train_on_batch(z_noise, valid_labels)

        # Salva le statistiche di training
        history['d_loss'].append(d_loss)
        history['g_loss'].append(g_loss)

        if epoch % 100 == 0:
            print(f"Epoch {epoch}, D_loss: {d_loss:.4f}, G_loss: {g_loss:.4f}")

            # Monitoraggio visivo occasionale della distribuzione degli score
            if epoch % 500 == 0 and epoch > 0:
                # Calcola gli score su un piccolo sottoinsieme
                subset_size = min(100, len(X_cross_val))
                subset_indices = np.random.choice(len(X_cross_val), subset_size, replace=False)
                subset_X = X_cross_val[subset_indices]
                subset_y = y_cross_val[subset_indices]

                scores = np.array([compute_anomaly_score(x) for x in subset_X])

                plt.figure(figsize=(8, 6))
                plt.hist(scores[subset_y == 0], bins=20, alpha=0.5, label="Normal (0)")
                plt.hist(scores[subset_y == 1], bins=20, alpha=0.5, label="Anomaly (1)")
                plt.xlabel("Anomaly score")
                plt.ylabel("Frequency")
                plt.legend()
                plt.title(f"GAN Anomaly Detection - Epoch {epoch}")
                plt.show()

    # Plot della loss history
    plt.figure(figsize=(10, 6))
    plt.plot(history['d_loss'], label='Discriminator Loss')
    plt.plot(history['g_loss'], label='Generator Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('GAN Training Loss')
    plt.grid(True)
    plt.show()

    return history

# --- Funzione migliorata per calcolare l'anomaly score ---
def compute_anomaly_score(x, n_attempts=15, alpha=0.6):  # Aumentato n_attempts e modificato alpha
    x = np.expand_dims(x, 0)

    # Genera campioni con noise pi variegato
    z_samples = []
    for _ in range(n_attempts // 3):
        # Noise standard
        z_samples.append(np.random.normal(0, 1, (1, LATENT_DIM)))
        # Noise con varianza minore
        z_samples.append(np.random.normal(0, 0.7, (1, LATENT_DIM)))
        # Noise con varianza maggiore
        z_samples.append(np.random.normal(0, 1.3, (1, LATENT_DIM)))

    z_samples = np.vstack(z_samples)

    # Genera esempi dal rumore
    x_gen_samples = generator.predict(z_samples, verbose=0)

    # Calcola errore di ricostruzione usando MSE
    rec_errors = np.mean(np.square(x_gen_samples - x), axis=1)

    # Score del discriminatore
    disc_scores = discriminator.predict(x, verbose=0).flatten()[0]

    # Combina i punteggi
    # Valori bassi di disc_score indicano anomalie, quindi usiamo (1-disc_score)
    combined_scores = alpha * rec_errors + (1 - alpha) * (1 - disc_scores)

    return np.min(combined_scores)

# --- Addestra il modello ---
history = train_gan(epochs=5000, batch_size=128)

# --- Valutazione su validation set ---
print("Calcolando gli anomaly score sul validation set...")
scores_cv = np.array([compute_anomaly_score(x) for x in X_cross_val])

# --- Ricerca della soglia ottimale ---
print("\nRicerca della soglia ottimale...")
print("Thresh\tPrecision\tRecall\tF1\t\tAUC")
best_f1 = 0
best_thresh = 0
min_recall = 0.75  # Vincolo di recall minimo

# Calcola ROC AUC
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_cross_val, scores_cv)
roc_auc = auc(fpr, tpr)
print(f"ROC AUC: {roc_auc:.4f}")

# Plotta ROC curve
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

for thresh in np.linspace(np.min(scores_cv), np.max(scores_cv), 100):
    y_pred_cv = (scores_cv > thresh).astype(int)
    prec = precision_score(y_cross_val, y_pred_cv)
    rec = recall_score(y_cross_val, y_pred_cv)
    f1_cv = f1_score(y_cross_val, y_pred_cv)
    print(f"{thresh:.4f}\t{prec:.3f}\t\t{rec:.3f}\t{f1_cv:.3f}")
    if rec >= min_recall and f1_cv > best_f1:
        best_f1 = f1_cv
        best_thresh = thresh

print(f"\nMigliore soglia con recall >= {min_recall}: {best_thresh:.4f}, F1={best_f1:.4f}")

# --- Valutazione finale sul test set ---
print("\nCalcolando gli anomaly score sul test set...")
scores_test = np.array([compute_anomaly_score(x) for x in X_test])
y_pred_test = (scores_test > best_thresh).astype(int)

cm = confusion_matrix(y_test, y_pred_test)
precision = precision_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
f1 = f1_score(y_test, y_pred_test)
accuracy = accuracy_score(y_test, y_pred_test)

print("\n--- Metriche di valutazione finali sul Test Set ---")
print(f"Confusion Matrix:\n{cm}")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")

# --- Plot della distribuzione degli score ---
plt.figure(figsize=(12, 8))
plt.hist(scores_test[y_test == 0], bins=50, alpha=0.5, label="Normal (0)", density=True)
plt.hist(scores_test[y_test == 1], bins=50, alpha=0.5, label="Anomaly (1)", density=True)
plt.axvline(best_thresh, color='r', linestyle='--', label=f'Threshold {best_thresh:.3f}')
plt.xlabel("Anomaly score")
plt.ylabel("Density")
plt.legend()
plt.title("GAN-based Anomaly Detection - Test Scores Distribution")
plt.grid(True)
plt.show()

# --- Visualizzazione dettagliata della confusion matrix ---
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal', 'Anomaly'],
            yticklabels=['Normal', 'Anomaly'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 8]

# --- Dataset ---
X = X_df.values

# --- Shuffle e split ---
X_shuffled, y_shuffled = shuffle(X, y, random_state=42)

X_normal = X_shuffled[y_shuffled == 0]
X_anomaly = X_shuffled[y_shuffled == 1]

n_normal = X_normal.shape[0]
n_anomaly = X_anomaly.shape[0]

train_size = int(0.8 * n_normal)
cv_normal_size = int(0.1 * n_normal)
cv_anomaly_size = int(0.5 * n_anomaly)

X_train_raw = X_normal[:train_size]  # solo normali
X_cv_normal = X_normal[train_size:train_size + cv_normal_size]
X_cv_anomaly = X_anomaly[:cv_anomaly_size]
X_cross_val_raw = np.vstack((X_cv_normal, X_cv_anomaly))
y_cross_val = np.hstack((np.zeros(cv_normal_size), np.ones(cv_anomaly_size)))

X_test_normal = X_normal[train_size + cv_normal_size:]
X_test_anomaly = X_anomaly[cv_anomaly_size:]
X_test_raw = np.vstack((X_test_normal, X_test_anomaly))
y_test = np.hstack((np.zeros(len(X_test_normal)), np.ones(len(X_test_anomaly))))

# --- StandardScaler invece di MinMaxScaler ---
scaler = StandardScaler()  # Cambiato a StandardScaler
X_train = scaler.fit_transform(X_train_raw)
X_cross_val = scaler.transform(X_cross_val_raw)
X_test = scaler.transform(X_test_raw)

print(f"Training set size: {X_train.shape[0]} (all normal)")
print(f"Cross-validation set size: {X_cross_val.shape[0]} ({cv_normal_size} normal, {cv_anomaly_size} anomalies)")
print(f"Test set size: {X_test.shape[0]} ({len(X_test_normal)} normal, {len(X_test_anomaly)} anomalies)")

# --- Iperparametri ---
LATENT_DIM = 20  # Riportato a un valore pi moderato
INPUT_DIM = X_train.shape[1]
REG_STRENGTH = 1e-5  # L2 regularization strength

# --- Autoencoder (approccio pi stabile per anomaly detection) ---
def build_autoencoder():
    # Encoder
    encoder_inputs = layers.Input(shape=(INPUT_DIM,))

    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(REG_STRENGTH))(encoder_inputs)
    x = layers.BatchNormalization()(x)

    x = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)

    # Latent representation
    latent = layers.Dense(LATENT_DIM, activation='relu', name='latent_layer')(x)

    # Decoder
    x = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(REG_STRENGTH))(latent)
    x = layers.BatchNormalization()(x)

    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)

    decoder_outputs = layers.Dense(INPUT_DIM, activation='linear')(x)

    # Modello completo
    autoencoder = models.Model(encoder_inputs, decoder_outputs, name='autoencoder')

    # Modello encoder separato
    encoder = models.Model(encoder_inputs, latent, name='encoder')

    return autoencoder, encoder

autoencoder, encoder = build_autoencoder()

# --- Compila e mostra il modello ---
autoencoder.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse'
)

autoencoder.summary()

# --- Callbacks ---
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    verbose=1,
    restore_best_weights=True
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=0.00001,
    verbose=1
)

# --- Addestra l'autoencoder solo su dati normali ---
history = autoencoder.fit(
    X_train, X_train,  # Input = Target (ricostruzione)
    epochs=100,
    batch_size=64,
    validation_split=0.1,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# --- Plot della loss durante l'addestramento ---
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Autoencoder Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# --- Funzione per calcolare l'anomaly score ---
def compute_anomaly_score(x):
    # Ricostruzione
    x_reconstructed = autoencoder.predict(np.array([x]), verbose=0)[0]

    # Errore di ricostruzione (MSE)
    reconstruction_error = np.mean(np.square(x - x_reconstructed))

    return reconstruction_error

# --- Calcola gli score su validation e test set ---
print("Calcolando gli anomaly score sul validation set...")
scores_cv = np.array([compute_anomaly_score(x) for x in X_cross_val])

print("Calcolando gli anomaly score sul test set...")
scores_test = np.array([compute_anomaly_score(x) for x in X_test])

# --- Visualizza la distribuzione degli score sul validation set ---
plt.figure(figsize=(10, 6))
plt.hist(scores_cv[y_cross_val == 0], bins=30, alpha=0.7, label='Normal', density=True)
plt.hist(scores_cv[y_cross_val == 1], bins=30, alpha=0.7, label='Anomaly', density=True)
plt.title('Distribution of Reconstruction Errors on Validation Set')
plt.xlabel('Reconstruction Error')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# --- Calcola ROC AUC ---
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_cross_val, scores_cv)
roc_auc = auc(fpr, tpr)
print(f"ROC AUC on validation set: {roc_auc:.4f}")

# Plotta ROC curve
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# --- Ricerca della soglia ottimale ---
print("\nRicerca della soglia ottimale...")
print("Thresh\tPrecision\tRecall\tF1\tAccuracy")

best_f1 = 0
best_thresh = 0
best_metrics = None

# Usa un approccio pi granulare per trovare la soglia ottimale
for thresh in np.linspace(np.percentile(scores_cv, 10), np.percentile(scores_cv, 99), 100):
    y_pred_cv = (scores_cv > thresh).astype(int)
    prec = precision_score(y_cross_val, y_pred_cv)
    rec = recall_score(y_cross_val, y_pred_cv)
    f1_cv = f1_score(y_cross_val, y_pred_cv)
    acc = accuracy_score(y_cross_val, y_pred_cv)

    # Vogliamo un buon equilibrio tra F1 e accuracy
    f1_adjusted = f1_cv

    if f1_adjusted > best_f1:
        best_f1 = f1_adjusted
        best_thresh = thresh
        best_metrics = (prec, rec, f1_cv, acc)

    if thresh % (np.percentile(scores_cv, 99) - np.percentile(scores_cv, 10)) / 20 < (np.percentile(scores_cv, 99) - np.percentile(scores_cv, 10)) / 100:
        print(f"{thresh:.4f}\t{prec:.3f}\t\t{rec:.3f}\t{f1_cv:.3f}\t{acc:.3f}")

prec, rec, f1_cv, acc = best_metrics
print(f"\nMigliore soglia: {best_thresh:.4f}")
print(f"Precision: {prec:.3f}, Recall: {rec:.3f}, F1: {f1_cv:.3f}, Accuracy: {acc:.3f}")

# --- Valutazione finale sul test set ---
y_pred_test = (scores_test > best_thresh).astype(int)

cm = confusion_matrix(y_test, y_pred_test)
precision = precision_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
f1 = f1_score(y_test, y_pred_test)
accuracy = accuracy_score(y_test, y_pred_test)

print("\n--- Metriche di valutazione finali sul Test Set ---")
print(f"Confusion Matrix:\n{cm}")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")

# --- Plot della distribuzione degli score sul test set ---
plt.figure(figsize=(10, 6))
plt.hist(scores_test[y_test == 0], bins=30, alpha=0.7, label='Normal', density=True)
plt.hist(scores_test[y_test == 1], bins=30, alpha=0.7, label='Anomaly', density=True)
plt.axvline(best_thresh, color='r', linestyle='--', label=f'Threshold {best_thresh:.4f}')
plt.title('Distribution of Reconstruction Errors on Test Set')
plt.xlabel('Reconstruction Error')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# --- Visualizzazione dettagliata della confusion matrix ---
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal', 'Anomaly'],
            yticklabels=['Normal', 'Anomaly'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# --- Feature importance basata su autoencoder ---
# Calcola l'importanza delle feature basata sull'errore di ricostruzione per feature
def feature_importance():
    # Prendi un campione del test set
    sample_size = min(500, X_test.shape[0])
    sample_indices = np.random.choice(X_test.shape[0], sample_size, replace=False)
    X_sample = X_test[sample_indices]

    # Ricostruisci
    X_reconstructed = autoencoder.predict(X_sample)

    # Calcola l'errore per feature
    feature_errors = np.mean(np.abs(X_sample - X_reconstructed), axis=0)

    # Normalizza
    feature_importance = feature_errors / np.sum(feature_errors)

    return feature_importance

# Calcola e visualizza l'importanza delle feature
feature_imp = feature_importance()

plt.figure(figsize=(12, 6))
plt.bar(range(len(feature_imp)), feature_imp)
plt.title('Feature Importance Based on Reconstruction Error')
plt.xlabel('Feature Index')
plt.ylabel('Importance')
plt.grid(True, axis='y')
plt.show()

# Mostra le top 10 feature pi importanti
top_indices = np.argsort(feature_imp)[-10:]
print("\nTop 10 feature pi importanti (indice, importanza):")
for idx in reversed(top_indices):
    print(f"Feature {idx}: {feature_imp[idx]:.4f}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 8]

# --- Dataset ---
X = X_df.values

# --- Shuffle e split ---
X_shuffled, y_shuffled = shuffle(X, y, random_state=42)

X_normal = X_shuffled[y_shuffled == 0]
X_anomaly = X_shuffled[y_shuffled == 1]

n_normal = X_normal.shape[0]
n_anomaly = X_anomaly.shape[0]

train_size = int(0.8 * n_normal)
cv_normal_size = int(0.1 * n_normal)
cv_anomaly_size = int(0.5 * n_anomaly)

X_train_raw = X_normal[:train_size]  # solo normali
X_cv_normal = X_normal[train_size:train_size + cv_normal_size]
X_cv_anomaly = X_anomaly[:cv_anomaly_size]
X_cross_val_raw = np.vstack((X_cv_normal, X_cv_anomaly))
y_cross_val = np.hstack((np.zeros(cv_normal_size), np.ones(cv_anomaly_size)))

X_test_normal = X_normal[train_size + cv_normal_size:]
X_test_anomaly = X_anomaly[cv_anomaly_size:]
X_test_raw = np.vstack((X_test_normal, X_test_anomaly))
y_test = np.hstack((np.zeros(len(X_test_normal)), np.ones(len(X_test_anomaly))))

# --- StandardScaler ---
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_raw)
X_cross_val = scaler.transform(X_cross_val_raw)
X_test = scaler.transform(X_test_raw)

print(f"Training set size: {X_train.shape[0]} (all normal)")
print(f"Cross-validation set size: {X_cross_val.shape[0]} ({cv_normal_size} normal, {cv_anomaly_size} anomalies)")
print(f"Test set size: {X_test.shape[0]} ({len(X_test_normal)} normal, {len(X_test_anomaly)} anomalies)")

# --- Iperparametri ---
LATENT_DIM = 8  # Ridotto a 8 come richiesto
INPUT_DIM = X_train.shape[1]
REG_STRENGTH = 1e-5
ACTIVATION = 'selu'  # SeLU per self-normalizing networks

# --- Autoencoder con skip connection ---
def build_deep_autoencoder_with_skip():
    # Input layer
    inputs = layers.Input(shape=(INPUT_DIM,))

    # Encoder
    # Primo blocco
    x = layers.Dense(256, activation=ACTIVATION, kernel_regularizer=regularizers.l2(REG_STRENGTH))(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.1)(x)
    skip1 = x  # Salvataggio per skip connection

    # Secondo blocco
    x = layers.Dense(128, activation=ACTIVATION, kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.1)(x)
    skip2 = x  # Salvataggio per skip connection

    # Terzo blocco
    x = layers.Dense(64, activation=ACTIVATION, kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.1)(x)
    skip3 = x  # Salvataggio per skip connection

    # Quarto blocco
    x = layers.Dense(32, activation=ACTIVATION, kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.1)(x)

    # Bottleneck (latent space)
    latent = layers.Dense(LATENT_DIM, activation=ACTIVATION, name='latent_layer',
                       kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)

    # Decoder
    # Primo blocco di decodifica
    x = layers.Dense(32, activation=ACTIVATION, kernel_regularizer=regularizers.l2(REG_STRENGTH))(latent)
    x = layers.BatchNormalization()(x)

    # Secondo blocco con skip connection
    x = layers.Dense(64, activation=ACTIVATION, kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Add()([x, skip3])  # Skip connection

    # Terzo blocco con skip connection
    x = layers.Dense(128, activation=ACTIVATION, kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Add()([x, skip2])  # Skip connection

    # Quarto blocco con skip connection
    x = layers.Dense(256, activation=ACTIVATION, kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Add()([x, skip1])  # Skip connection

    # Output layer
    outputs = layers.Dense(INPUT_DIM, activation='linear')(x)

    # Modello completo
    autoencoder = models.Model(inputs, outputs, name='deep_autoencoder_with_skip')

    # Modello encoder separato per estrarre la rappresentazione latente
    encoder = models.Model(inputs, latent, name='encoder')

    return autoencoder, encoder

# --- Callback per monitorare l'errore di ricostruzione ---
class ReconstructionMonitor(tf.keras.callbacks.Callback):
    def __init__(self, validation_data, interval=5):
        super(ReconstructionMonitor, self).__init__()
        self.validation_data = validation_data
        self.interval = interval

    def on_epoch_end(self, epoch, logs=None):
        if epoch % self.interval == 0:
            x_val, _ = self.validation_data
            x_val_pred = self.model.predict(x_val)
            mse = np.mean(np.square(x_val - x_val_pred))
            print(f'\nEpoch {epoch}: validation reconstruction MSE: {mse:.6f}')

# --- Costruisci il modello ---
autoencoder, encoder = build_deep_autoencoder_with_skip()

# --- Compila ---
autoencoder.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse',
    metrics=['mae']  # Aggiungi Mean Absolute Error come metrica
)

# --- Mostra l'architettura ---
autoencoder.summary()

# --- Callbacks ---
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=30,
    verbose=1,
    restore_best_weights=True
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=15,
    min_lr=0.00001,
    verbose=1
)

reconstruction_monitor = ReconstructionMonitor(
    validation_data=(X_train[:1000], X_train[:1000]),
    interval=10
)

# --- Addestra l'autoencoder ---
history = autoencoder.fit(
    X_train, X_train,
    epochs=1000,
    batch_size=64,
    validation_split=0.1,
    callbacks=[early_stopping, reduce_lr, reconstruction_monitor],
    verbose=1
)

# --- Plot della loss durante l'addestramento ---
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Autoencoder Training Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True)
plt.show()

# --- Visualizza l'andamento dell'errore MAE ---
plt.figure(figsize=(10, 6))
plt.plot(history.history['mae'], label='Training MAE')
plt.plot(history.history['val_mae'], label='Validation MAE')
plt.title('Autoencoder Mean Absolute Error')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.grid(True)
plt.show()

# --- Funzione per calcolare l'anomaly score ---
def compute_anomaly_score(x, mode='combined'):
    # Ricostruzione
    x_reconstructed = autoencoder.predict(np.array([x]), verbose=0)[0]

    # Errore di ricostruzione (MSE)
    reconstruction_error = np.mean(np.square(x - x_reconstructed))

    if mode == 'combined':
        # Ottieni la rappresentazione latente
        latent_representation = encoder.predict(np.array([x]), verbose=0)[0]

        # Calcola un punteggio di anomalia basato sulla rappresentazione latente
        # (deviazione dalla media delle rappresentazioni latenti normali)
        latent_deviation = np.sum(np.square(latent_representation))

        # Combina gli score (puoi regolare il peso )
        alpha = 0.7
        combined_score = alpha * reconstruction_error + (1 - alpha) * latent_deviation
        return combined_score
    else:
        return reconstruction_error

# --- Calcola gli score su validation e test set ---
print("Calcolando gli anomaly score sul validation set...")
scores_cv = np.array([compute_anomaly_score(x) for x in X_cross_val])

print("Calcolando gli anomaly score sul test set...")
scores_test = np.array([compute_anomaly_score(x) for x in X_test])

# --- Visualizza la distribuzione degli score sul validation set ---
plt.figure(figsize=(10, 6))
plt.hist(scores_cv[y_cross_val == 0], bins=30, alpha=0.7, label='Normal', density=True)
plt.hist(scores_cv[y_cross_val == 1], bins=30, alpha=0.7, label='Anomaly', density=True)
plt.title('Distribution of Anomaly Scores on Validation Set')
plt.xlabel('Anomaly Score')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# --- Calcola ROC AUC ---
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score
fpr, tpr, thresholds = roc_curve(y_cross_val, scores_cv)
roc_auc = auc(fpr, tpr)
print(f"ROC AUC on validation set: {roc_auc:.4f}")

# Calcola anche Precision-Recall AUC (utile per dataset sbilanciati)
precision_curve, recall_curve, _ = precision_recall_curve(y_cross_val, scores_cv)
pr_auc = average_precision_score(y_cross_val, scores_cv)
print(f"PR AUC on validation set: {pr_auc:.4f}")

# Plotta ROC curve
plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.grid(True)

# Plotta PR curve
plt.subplot(1, 2, 2)
plt.plot(recall_curve, precision_curve, color='blue', lw=2, label=f'PR curve (area = {pr_auc:.3f})')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")
plt.grid(True)

plt.tight_layout()
plt.show()

# --- Ricerca della soglia ottimale ---
print("\nRicerca della soglia ottimale...")
print("Thresh\tPrecision\tRecall\tF1\tAccuracy")

best_f1 = 0
best_thresh = 0
best_metrics = None

# Usa un approccio pi granulare per trovare la soglia ottimale
for thresh in np.linspace(np.percentile(scores_cv, 10), np.percentile(scores_cv, 99), 100):
    y_pred_cv = (scores_cv > thresh).astype(int)
    prec = precision_score(y_cross_val, y_pred_cv)
    rec = recall_score(y_cross_val, y_pred_cv)
    f1_cv = f1_score(y_cross_val, y_pred_cv)
    acc = accuracy_score(y_cross_val, y_pred_cv)

    # Vogliamo massimizzare F1
    if f1_cv > best_f1:
        best_f1 = f1_cv
        best_thresh = thresh
        best_metrics = (prec, rec, f1_cv, acc)

    if thresh % (np.percentile(scores_cv, 99) - np.percentile(scores_cv, 10)) / 10 < (np.percentile(scores_cv, 99) - np.percentile(scores_cv, 10)) / 100:
        print(f"{thresh:.4f}\t{prec:.3f}\t\t{rec:.3f}\t{f1_cv:.3f}\t{acc:.3f}")

prec, rec, f1_cv, acc = best_metrics
print(f"\nMigliore soglia: {best_thresh:.4f}")
print(f"Precision: {prec:.3f}, Recall: {rec:.3f}, F1: {f1_cv:.3f}, Accuracy: {acc:.3f}")

# --- Valutazione finale sul test set ---
y_pred_test = (scores_test > best_thresh).astype(int)

cm = confusion_matrix(y_test, y_pred_test)
precision = precision_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
f1 = f1_score(y_test, y_pred_test)
accuracy = accuracy_score(y_test, y_pred_test)

print("\n--- Metriche di valutazione finali sul Test Set ---")
print(f"Confusion Matrix:\n{cm}")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")

# --- Plot della distribuzione degli score sul test set ---
plt.figure(figsize=(10, 6))
plt.hist(scores_test[y_test == 0], bins=30, alpha=0.7, label='Normal', density=True)
plt.hist(scores_test[y_test == 1], bins=30, alpha=0.7, label='Anomaly', density=True)
plt.axvline(best_thresh, color='r', linestyle='--', label=f'Threshold {best_thresh:.4f}')
plt.title('Distribution of Anomaly Scores on Test Set')
plt.xlabel('Anomaly Score')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# --- Visualizzazione dettagliata della confusion matrix ---
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal', 'Anomaly'],
            yticklabels=['Normal', 'Anomaly'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# --- Visualizzazione esempi di ricostruzione ---
def plot_reconstructions(n_samples=5):
    # Prendi alcuni esempi casuali di dati normali e anomali
    normal_idx = np.random.choice(np.where(y_test == 0)[0], n_samples)
    anomaly_idx = np.random.choice(np.where(y_test == 1)[0], n_samples)

    fig, axes = plt.subplots(2, n_samples, figsize=(15, 6))

    for i, idx in enumerate(normal_idx):
        # Dati originali
        original = X_test[idx]
        # Ricostruzione
        reconstructed = autoencoder.predict(np.array([original]), verbose=0)[0]
        # Errore
        error = np.abs(original - reconstructed)

        # Plot dati originali vs ricostruiti (primi 20 features)
        axes[0, i].plot(range(20), original[:20], 'b-', label='Original')
        axes[0, i].plot(range(20), reconstructed[:20], 'r--', label='Reconstructed')
        axes[0, i].set_title(f'Normal Sample {i+1}')
        if i == 0:
            axes[0, i].legend()

    for i, idx in enumerate(anomaly_idx):
        # Dati originali
        original = X_test[idx]
        # Ricostruzione
        reconstructed = autoencoder.predict(np.array([original]), verbose=0)[0]
        # Errore
        error = np.abs(original - reconstructed)

        # Plot dati originali vs ricostruiti (primi 20 features)
        axes[1, i].plot(range(20), original[:20], 'b-', label='Original')
        axes[1, i].plot(range(20), reconstructed[:20], 'r--', label='Reconstructed')
        axes[1, i].set_title(f'Anomaly Sample {i+1}')
        if i == 0:
            axes[1, i].legend()

    plt.tight_layout()
    plt.show()

# Visualizza alcuni esempi di ricostruzione
plot_reconstructions(5)

# --- Feature importance basata su reconstruction error ---
def feature_importance():
    # Prendi un campione del test set
    normal_samples = X_test[y_test == 0]
    anomaly_samples = X_test[y_test == 1]

    # Ricostruisci
    normal_reconstructed = autoencoder.predict(normal_samples)
    anomaly_reconstructed = autoencoder.predict(anomaly_samples)

    # Calcola l'errore per feature
    normal_errors = np.mean(np.abs(normal_samples - normal_reconstructed), axis=0)
    anomaly_errors = np.mean(np.abs(anomaly_samples - anomaly_reconstructed), axis=0)

    # Differenza negli errori
    error_diff = anomaly_errors - normal_errors

    # Normalizza
    feature_importance = error_diff / np.sum(np.abs(error_diff))

    return feature_importance

# Calcola e visualizza l'importanza delle feature
feature_imp = feature_importance()

plt.figure(figsize=(12, 6))
plt.bar(range(len(feature_imp)), feature_imp)
plt.title('Feature Importance Based on Reconstruction Error Difference')
plt.xlabel('Feature Index')
plt.ylabel('Importance')
plt.grid(True, axis='y')
plt.show()

# Mostra le top 10 feature pi importanti
top_indices = np.argsort(np.abs(feature_imp))[-10:]
print("\nTop 10 feature pi importanti (indice, importanza):")
for idx in reversed(top_indices):
    print(f"Feature {idx}: {feature_imp[idx]:.4f}")

# --- Visualizzazione dello spazio latente ---
from sklearn.manifold import TSNE

# Estrai la rappresentazione latente per tutti i dati di test
latent_test = encoder.predict(X_test)

# Riduci la dimensionalit per la visualizzazione
tsne = TSNE(n_components=2, random_state=42)
latent_2d = tsne.fit_transform(latent_test)

# Visualizza lo spazio latente
plt.figure(figsize=(10, 8))
plt.scatter(latent_2d[y_test == 0, 0], latent_2d[y_test == 0, 1], c='blue', alpha=0.5, label='Normal')
plt.scatter(latent_2d[y_test == 1, 0], latent_2d[y_test == 1, 1], c='red', alpha=0.5, label='Anomaly')
plt.title('t-SNE visualization of latent space')
plt.xlabel('t-SNE dimension 1')
plt.ylabel('t-SNE dimension 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 8]

# --- Dataset ---
X = X_df.values

# --- Shuffle e split ---
X_shuffled, y_shuffled = shuffle(X, y, random_state=42)

X_normal = X_shuffled[y_shuffled == 0]
X_anomaly = X_shuffled[y_shuffled == 1]

n_normal = X_normal.shape[0]
n_anomaly = X_anomaly.shape[0]

train_size = int(0.8 * n_normal)
cv_normal_size = int(0.1 * n_normal)
cv_anomaly_size = int(0.5 * n_anomaly)

X_train_raw = X_normal[:train_size]  # solo normali
X_cv_normal = X_normal[train_size:train_size + cv_normal_size]
X_cv_anomaly = X_anomaly[:cv_anomaly_size]
X_cross_val_raw = np.vstack((X_cv_normal, X_cv_anomaly))
y_cross_val = np.hstack((np.zeros(cv_normal_size), np.ones(cv_anomaly_size)))

X_test_normal = X_normal[train_size + cv_normal_size:]
X_test_anomaly = X_anomaly[cv_anomaly_size:]
X_test_raw = np.vstack((X_test_normal, X_test_anomaly))
y_test = np.hstack((np.zeros(len(X_test_normal)), np.ones(len(X_test_anomaly))))

# --- StandardScaler invece di MinMaxScaler ---
scaler = StandardScaler()  # Cambiato a StandardScaler
X_train = scaler.fit_transform(X_train_raw)
X_cross_val = scaler.transform(X_cross_val_raw)
X_test = scaler.transform(X_test_raw)

print(f"Training set size: {X_train.shape[0]} (all normal)")
print(f"Cross-validation set size: {X_cross_val.shape[0]} ({cv_normal_size} normal, {cv_anomaly_size} anomalies)")
print(f"Test set size: {X_test.shape[0]} ({len(X_test_normal)} normal, {len(X_test_anomaly)} anomalies)")

# --- Iperparametri ---
LATENT_DIM = 20  # Riportato a un valore pi moderato
INPUT_DIM = X_train.shape[1]
REG_STRENGTH = 1e-5  # L2 regularization strength

# --- Autoencoder (approccio pi stabile per anomaly detection) ---
def build_autoencoder():
    # Encoder
    encoder_inputs = layers.Input(shape=(INPUT_DIM,))

    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(REG_STRENGTH))(encoder_inputs)
    x = layers.BatchNormalization()(x)

    x = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)

    # Latent representation
    latent = layers.Dense(LATENT_DIM, activation='relu', name='latent_layer')(x)

    # Decoder
    x = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(REG_STRENGTH))(latent)
    x = layers.BatchNormalization()(x)

    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(REG_STRENGTH))(x)
    x = layers.BatchNormalization()(x)

    decoder_outputs = layers.Dense(INPUT_DIM, activation='linear')(x)

    # Modello completo
    autoencoder = models.Model(encoder_inputs, decoder_outputs, name='autoencoder')

    # Modello encoder separato
    encoder = models.Model(encoder_inputs, latent, name='encoder')

    return autoencoder, encoder

autoencoder, encoder = build_autoencoder()

# --- Compila e mostra il modello ---
autoencoder.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='mse'
)

autoencoder.summary()

# --- Callbacks ---
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=20,
    verbose=1,
    restore_best_weights=True
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=15,
    min_lr=0.00001,
    verbose=1
)

# --- Addestra l'autoencoder solo su dati normali ---
history = autoencoder.fit(
    X_train, X_train,  # Input = Target (ricostruzione)
    epochs=500,
    batch_size=64,
    validation_split=0.1,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# --- Plot della loss durante l'addestramento ---
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Autoencoder Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# --- Funzione per calcolare l'anomaly score ---
def compute_anomaly_score(x):
    # Ricostruzione
    x_reconstructed = autoencoder.predict(np.array([x]), verbose=0)[0]

    # Errore di ricostruzione (MSE)
    reconstruction_error = np.mean(np.square(x - x_reconstructed))

    return reconstruction_error

# --- Calcola gli score su validation e test set ---
print("Calcolando gli anomaly score sul validation set...")
scores_cv = np.array([compute_anomaly_score(x) for x in X_cross_val])

print("Calcolando gli anomaly score sul test set...")
scores_test = np.array([compute_anomaly_score(x) for x in X_test])

# --- Visualizza la distribuzione degli score sul validation set ---
plt.figure(figsize=(10, 6))
plt.hist(scores_cv[y_cross_val == 0], bins=30, alpha=0.7, label='Normal', density=True)
plt.hist(scores_cv[y_cross_val == 1], bins=30, alpha=0.7, label='Anomaly', density=True)
plt.title('Distribution of Reconstruction Errors on Validation Set')
plt.xlabel('Reconstruction Error')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# --- Calcola ROC AUC ---
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_cross_val, scores_cv)
roc_auc = auc(fpr, tpr)
print(f"ROC AUC on validation set: {roc_auc:.4f}")

# Plotta ROC curve
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# --- Ricerca della soglia ottimale ---
print("\nRicerca della soglia ottimale...")
print("Thresh\tPrecision\tRecall\tF1\tAccuracy")

best_f1 = 0
best_thresh = 0
best_metrics = None

# Usa un approccio pi granulare per trovare la soglia ottimale
for thresh in np.linspace(np.percentile(scores_cv, 10), np.percentile(scores_cv, 99), 100):
    y_pred_cv = (scores_cv > thresh).astype(int)
    prec = precision_score(y_cross_val, y_pred_cv)
    rec = recall_score(y_cross_val, y_pred_cv)
    f1_cv = f1_score(y_cross_val, y_pred_cv)
    acc = accuracy_score(y_cross_val, y_pred_cv)

    # Vogliamo un buon equilibrio tra F1 e accuracy
    f1_adjusted = f1_cv

    if f1_adjusted > best_f1:
        best_f1 = f1_adjusted
        best_thresh = thresh
        best_metrics = (prec, rec, f1_cv, acc)

    if thresh % (np.percentile(scores_cv, 99) - np.percentile(scores_cv, 10)) / 20 < (np.percentile(scores_cv, 99) - np.percentile(scores_cv, 10)) / 100:
        print(f"{thresh:.4f}\t{prec:.3f}\t\t{rec:.3f}\t{f1_cv:.3f}\t{acc:.3f}")

prec, rec, f1_cv, acc = best_metrics
print(f"\nMigliore soglia: {best_thresh:.4f}")
print(f"Precision: {prec:.3f}, Recall: {rec:.3f}, F1: {f1_cv:.3f}, Accuracy: {acc:.3f}")

# --- Valutazione finale sul test set ---
y_pred_test = (scores_test > best_thresh).astype(int)

cm = confusion_matrix(y_test, y_pred_test)
precision = precision_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
f1 = f1_score(y_test, y_pred_test)
accuracy = accuracy_score(y_test, y_pred_test)

print("\n--- Metriche di valutazione finali sul Test Set ---")
print(f"Confusion Matrix:\n{cm}")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")

# --- Plot della distribuzione degli score sul test set ---
plt.figure(figsize=(10, 6))
plt.hist(scores_test[y_test == 0], bins=30, alpha=0.7, label='Normal', density=True)
plt.hist(scores_test[y_test == 1], bins=30, alpha=0.7, label='Anomaly', density=True)
plt.axvline(best_thresh, color='r', linestyle='--', label=f'Threshold {best_thresh:.4f}')
plt.title('Distribution of Reconstruction Errors on Test Set')
plt.xlabel('Reconstruction Error')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# --- Visualizzazione dettagliata della confusion matrix ---
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal', 'Anomaly'],
            yticklabels=['Normal', 'Anomaly'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# --- Feature importance basata su autoencoder ---
# Calcola l'importanza delle feature basata sull'errore di ricostruzione per feature
def feature_importance():
    # Prendi un campione del test set
    sample_size = min(500, X_test.shape[0])
    sample_indices = np.random.choice(X_test.shape[0], sample_size, replace=False)
    X_sample = X_test[sample_indices]

    # Ricostruisci
    X_reconstructed = autoencoder.predict(X_sample)

    # Calcola l'errore per feature
    feature_errors = np.mean(np.abs(X_sample - X_reconstructed), axis=0)

    # Normalizza
    feature_importance = feature_errors / np.sum(feature_errors)

    return feature_importance

# Calcola e visualizza l'importanza delle feature
feature_imp = feature_importance()

plt.figure(figsize=(12, 6))
plt.bar(range(len(feature_imp)), feature_imp)
plt.title('Feature Importance Based on Reconstruction Error')
plt.xlabel('Feature Index')
plt.ylabel('Importance')
plt.grid(True, axis='y')
plt.show()

# Mostra le top 10 feature pi importanti
top_indices = np.argsort(feature_imp)[-10:]
print("\nTop 10 feature pi importanti (indice, importanza):")
for idx in reversed(top_indices):
    print(f"Feature {idx}: {feature_imp[idx]:.4f}")
```
